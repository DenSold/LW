2. Анализ данных о сжённых калориях людьми, посещающими спортзал (источник: https://www.kaggle.com/datasets/valakhorasani/gym-members-exercise-dataset)

```python
data = pd.read_csv("gym_members_exercise_tracking.csv")
data
# Извлечение столбца
bpm = data["Calories_Burned"].dropna()  # Убираем пропуски

# Построение гистограммы
plt.hist(bpm, bins=20, density=True, alpha=0.6, color="b", label="Калории")

# Оценка параметров нормального распределения
mu_1, std_1 = bpm.mean(), bpm.std()

# Создание кривой нормального распределения
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu_1, std_1)

# Наложение кривой нормального распределения
plt.plot(x, p, "k", linewidth=2, label="Нормальное распределение")

# Оформление графика
plt.title("Распределение сжённых калорий за тренировку")
plt.xlabel("Калории")
plt.ylabel("Плотность")
plt.legend()
plt.show()
```

```python
# 1. Математическое ожидание (среднее значение)
mean_value_2 = bpm.mean()
print(f"Математическое ожидание: {mean_value_2:.3f} калорий")

# 2. Дисперсия
variance_2 = bpm.var()
print(f"Дисперсия: {variance_2:.3f} калорий")

# 3. Асимметрия
skewness_2 = bpm.skew()
print(f"Асимметрия: {skewness_2:.3f}")

# 4. Эксцесс
kurtosis_2 = bpm.kurtosis()
print(f"Эксцесс: {kurtosis_2:.3f}")

# 5. Квантили уровня 0,05 и 0,95
quantile_05_2 = bpm.quantile(0.05)
quantile_95_2 = bpm.quantile(0.95)
print(f"Квантиль 0,05: {quantile_05_2:.3f} калорий")
print(f"Квантиль 0,95: {quantile_95_2:.3f} калорий")

# 2,5%-ная точка
quantile_025_2 = bpm.quantile(0.025)
print(f"2,5%-ная точка: {quantile_025_2:.3f} калорий")
```

- **Эксцесс:** Эксцесс -0.056 указывает на более «плоское» распределение по сравнению с нормальным, что говорит о менее выраженных «пиках» около среднего значения.

- **Квантиль 0.05:** Только 5% людей сожгли 448 калорий за тренировку, что отражает низшую границу значений для большинства выборки.

- **Квантиль 0.95:** Большинство людей (95%) сожгли 1378 калорий, и только 5% превышают этот показатель, что помогает оценить верхнюю границу роста.
