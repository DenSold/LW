# Импорт библиотек

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

# Загружаем исходные данные

В выбранном датасете представлен набор факторов, влияющих на успеваемость учащихся во время экзаменов. Она включает информацию об учебных привычках, посещаемости, участии родителей и других аспектах, влияющих на успеваемость.


Ссылка на датасет: https://www.kaggle.com/datasets/lainguyn123/student-performance-factors

```python
dtst = pd.read_csv('StudentPerformanceFactors.csv')
dtst.head()
```

# Нормальное распределение

Предположим, что близко к нормальному распределению будут данные о количестве часов обучения. Проверим это предположение с помощью гистограммы

```python
sns.histplot(data=dtst, x='Hours_Studied', bins=42, kde=True)
plt.show()
```

# Равномерное распределение

Предположим, что близко к равномерному распределению будут данные о прошлых оценках. Проверим это предположение с помощью гистограммы

```python
sns.histplot(data=dtst, x='Previous_Scores', bins=25, kde=True)
plt.show()
```

# Распределение со смещенной медианой

```python
sns.histplot(data=dtst, x='Tutoring_Sessions', bins=7)
plt.show()
print(f"Среднее = {dtst.Tutoring_Sessions.mean()}")
print(f"Медиана = {dtst.Tutoring_Sessions.median()}")
```

# Распределение с несмещенной медианой

```python
sns.histplot(data=dtst, x='Sleep_Hours', bins=7)
plt.show()
print(f"Среднее = {dtst.Sleep_Hours.mean()}")
print(f"Медиана = {dtst.Sleep_Hours.median()}")
```

# Описательные характеристики 

Среднее, стандартное отклонение, минимум, максимум, квантили

```python
dtst[["Hours_Studied", "Sleep_Hours"]].describe()
```

Мода

```python
print("Мода: ")
dtst[["Hours_Studied", "Sleep_Hours"]].mode()
```

# График рассеивания

```python
sns.scatterplot(data=dtst, y="Hours_Studied", x="Sleep_Hours")
plt.show()
```

# Круговая диаграмма

```python
data = dtst['Extracurricular_Activities'].value_counts()
plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Extracurricular Activities Distribution')
plt.axis('equal')
plt.show()
```

# Распределение с аномальными значениями

Каждая запись в датасете является набором показателей португальского красного вина "Vinho Verde". Был взят показатель pH (Кислотность). 

Ссылка на датасет: https://www.kaggle.com/datasets/uciml/red-wine-quality-cortez-et-al-2009

```python
dtst = pd.read_csv("winequality-red.csv")
plt.boxplot(dtst.pH)
plt.ylabel("Кислотность")
plt.show()
```

Аномальные значения могут быть связаны с тем, что в большинство партий данного вина были произведены в разных регионах с разной почвой и климатом, что привело к отклонениям в показателях в сравнении с оригиналом.
